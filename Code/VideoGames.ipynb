### Question 15: Find the average price of apps with in-app purchases vs. those without.

```python
df['In-app Purchases']=df['In-app Purchases'].apply(lambda x: float(x.split(",")[0]) if isinstance(x, str) and x != "0" else float(x))
```

```python
df['With_iap']=df['In-app Purchases']>0
avg_iap=df[df['With_iap']]['Price'].mean()
avg_without_iap=df[~df['With_iap']]['Price'].mean()
print(f" Average price of apps with in app purchases is {avg_iap}")
print(f" Average price of apps without in app purchases is {avg_without_iap}")
```

### Question 16: Are there any apps priced unusually high (outliers)?

```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Find high-priced apps
high_priced_apps = df[df["Price"] > upper_bound]
print(high_priced_apps[["Name", "Price"]].sort_values(by=['Price'],ascending=False).head(3))
```
